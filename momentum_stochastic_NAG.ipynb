```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad(x):
    return 2*x+10*np.cos(x)

def cost(x):
    return x**2+10*np.sin(x)
```

```python
np.random.seed(2)
def hoitu(w,grad):
    return np.linalg.norm(grad(w))/len(w) < 1e-3

def GD_momentum(w_init, eta, gamma):
    w = [w_init]
    v = [np.zeros_like(w_init)]
    for i in range(100000):
        v_new  = gamma * v[-1] + eta * grad(w[-1] - gamma*v[-1]) #them NAG
        w_new = w[-1] - v_new
        if hoitu(w_new, grad):
            break
        w.append(w_new)
        v.append(v_new)
    print (i)
    return w[-1]
s = cost(GD_momentum([[10]], 0.4, 0.6))
print(s)
```
